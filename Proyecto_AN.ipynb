## Data loading

### Subtask:

```python
import pandas as pd

try:
    df = pd.read_csv('datos.csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'datos.csv' not found. Please ensure the file exists in the current directory.")
    df = None  # Set df to None to indicate failure
except pd.errors.ParserError:
    print("Error: Unable to parse 'datos.csv'. Please check the file format.")
    df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
```

## Data cleaning

### Subtask:
Clean the data by handling inconsistencies and data type errors identified during data exploration.


**Reasoning**:
Convert date columns to datetime objects, drop the `departamento` and `cantFilas` columns due to lack of variability, and check for missing values after these operations.



```python
import pandas as pd

# Convert date columns to datetime objects
for col in ['fechaProceso', 'fechaUltimaActuacion']:
    try:
        df[col] = pd.to_datetime(df[col])
    except ValueError as e:
        print(f"Error converting {col}: {e}")
        # Handle the error, e.g., replace invalid dates with NaT
        df[col] = pd.to_datetime(df[col], errors='coerce')

# Drop columns with constant values
if len(df['departamento'].unique()) == 1:
    df = df.drop('departamento', axis=1)

if len(df['cantFilas'].unique()) == 1:
    df = df.drop('cantFilas', axis=1)

# Check for missing values again
print("\nMissing Values after data cleaning:\n", df.isnull().sum() / len(df) * 100)

display(df.head())
```
